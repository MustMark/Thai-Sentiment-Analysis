# Logistic Regression

```python
import json
import pandas as pd

with open("../../dataset/train_sentiment.json", "r", encoding="utf-8") as f:
    dataset = json.load(f)

df = pd.DataFrame(dataset)

texts = df["text"]
labels = df["sentiment"]
```

```python
from sklearn.model_selection import train_test_split
# 2. Split data
text_train, text_test, sentiment_train, sentiment_test = train_test_split(
    texts, labels, test_size=0.2, random_state=42
)
```

```python
import sys
import os
sys.path.append(os.path.abspath("../../"))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from models.preprocess import tokenizer

# 1. Convert text to numbers (TF-IDF)
vectorizer = TfidfVectorizer(
    tokenizer=tokenizer,
    token_pattern=None,
    ngram_range=(1,2),
    max_features=60000,
    min_df=2,
    sublinear_tf=True)

text_train_tfidf = vectorizer.fit_transform(text_train)
text_test_tfidf = vectorizer.transform(text_test)
# joblib.dump(vectorizer, "vectorizer.pkl")
```

```python
from sklearn.linear_model import LogisticRegression # type: ignore
import joblib

# 3. Create model
model = LogisticRegression()

# 4. Train model
model.fit(text_train_tfidf, sentiment_train)

joblib.dump((vectorizer, model), "model.pkl")
```

```python
from sklearn.metrics import accuracy_score, classification_report # type: ignore
# 5. Predict
predictions = model.predict(text_test_tfidf)

# 6. Evaluate
print("Accuracy:", accuracy_score(sentiment_test, predictions))
print(classification_report(sentiment_test, predictions))
```
